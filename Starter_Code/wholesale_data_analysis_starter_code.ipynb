## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
#   subtotal = unit_price * qty
#   add new column "subtotal"

df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
#   if order_weight >50, calculate shipping price at $7 * unit_weight
#   if order_weight <=50, $10 * unit weight
#   add new column "shipping_price"

df['order_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df.apply(lambda row: 7 * row['order_weight'] if row['order_weight'] > 50 else 10 * row['order_weight'], axis=1)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%


df['total_price'] = df['total_price'] = (df['subtotal'] + df['shipping_price']) * 1.0925
df['total_price'] = df['total_price'].round(2)
df.head()

```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
#   create new column "line_cost"
#   line_cost=(unit_cost+shipping_price)*qty

df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
#   create new column "profit"
#   profit = subtotal- line_cost

df['profit'] = df['total_price'] - df['line_cost']
df = df.round(2)
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
pd.set_option('display.max_colwidth', None)

top_clients_data = df[df['client_id'].isin(top_five_clients)]

total_price_sum_per_client = top_clients_data.groupby('client_id')['total_price'].sum().reset_index()

total_price_sum_per_client['total_price'] = total_price_sum_per_client['total_price'].astype(float)

total_price_sum_per_client['total_price'] = total_price_sum_per_client['total_price'].map('${:,.2f}'.format)

print(total_price_sum_per_client)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

summary_df = df[df['client_id'].isin(top_five_clients)]
summed_values = summary_df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'line_cost': 'sum',
    'profit': 'sum'
}).reset_index()

# filtered_orders
summed_values.head()

```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
# Define a function that converts a dollar amount to millions.
# Apply the currency_format_millions function to only the money columns. 
# Rename the columns to reflect the change in the money format. 



summed_values_formatted = summed_values.rename(columns={'client_id':"Client ID",
                                                        'qty':'Units',
                                                        'shipping_price':'Shipping (millions)',
                                                        'total_price':'Total Revenue (millions)',
                                                        'line_cost':'Total Cost (millions)',
                                                        'profit':'Total Profit (millions)'})
summed_values_formatted.head()

def convert_to_millions(x):
    return '${:,.2f}M'.format(x / 1000000)

money_columns = ['Shipping (millions)', 'Total Revenue (millions)', 'Total Cost (millions)', 'Total Profit (millions)']

summed_values_formatted[money_columns] = summed_values_formatted[money_columns].apply(lambda x: x.apply(convert_to_millions))

summed_values_formatted


```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
formatted_sorted = summed_values_formatted.sort_values('Total Profit (millions)', ascending=False)
formatted_sorted
```

#####   The averages for Total Revenue, Total Cost, and Total Profit are heavily weighted by the most profitable client, 24741;this needs to be considered in any subsequent analyses of the full data to ensure that reporting averages is not skewed  by the results of a single client.
#####   The Cost to Revenue ratio for client 24741 is significantly lower in comparison to the remaining top four clients; further analysis is reccomended to understand why they are so much more profitable in relation to the other top clients.
